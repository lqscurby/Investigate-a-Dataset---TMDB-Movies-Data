### Data Cleaning:

```python
# After discussing the structure of the data and any problems that need to be
#   cleaned, perform those cleaning steps in the second part of this section.

#deleting unused columns
del_col=['id','imdb_id','homepage','tagline', 'keywords', 'overview', 'production_companies']

#delete the columns
tmdb_data = tmdb_data.drop(del_col,1)
```

### Research Question 1: What is the profit for each movie?

```python
# Use this, and more code cells, to explore your data. Don't forget to add
#   Markdown cells to document your observations and findings.

#insert new column for profit value
tmdb_data.insert(3,'profit',tmdb_data['revenue']-tmdb_data['budget'])
tmdb_data.head(1)               
```

Research Question 2: What is the most and least profit.
The most profit is 

```python
#check mean profit of movies
tmdb_data.groupby('profit').mean()
tmdb_data.describe()
```

Research Question 3: Which movies with most and least profit.

```python
#define the function
def calculate(column):
    #high profit
    high=tmdb_data[column].idxmax()
    high_data=pd.DataFrame(tmdb_data.loc[high])
    
    #low profit
    low=tmdb_data[column].idxmin()
    low_data=pd.DataFrame(tmdb_data.loc[low])
    
    #collect data
    data=pd.concat([high_data, low_data], axis=1)
    
    return data

calculate ('profit')
    
```

As you can see the original title movie 'Avatar' has the most profit of 2544505847, and the movie 'The Warrior's Way' has the least profit of -413912431.
